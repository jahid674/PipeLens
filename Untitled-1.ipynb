```python
import pandas as pd
import matplotlib.pyplot as plt

df1 = pd.read_csv('historical_data/similarity/P1_F2_lr_sp_adult.csv')
df2 = pd.read_csv('historical_data/similarity/P1_F4_lr_sp_adult.csv')
df3 = pd.read_csv('historical_data/similarity/encoding_lr_sp_adult.csv')


col1 = df1.columns[-1]
col2 = df2.columns[-1]
col3 = df3.columns[-1]

plt.figure(figsize=(10, 6), dpi=300)
plt.plot(df1[col1].values, label='Outlier')
plt.plot(df2[col2].values, label='Fselection')
plt.plot(df3[col3].values, label='Encoding')

plt.title("Comparison of Similarity Metrics: [Imp, Norm, Model](P) vs Other Pipelines(F)")
plt.xlabel("Param Combinations")
plt.ylabel("Value")
plt.legend()
plt.tight_layout()
plt.show()
```

```python

df1 = pd.read_csv('historical_data/similarity/P1_F2_lr_sp_adult.csv')
df2 = pd.read_csv('historical_data/noise/full_exp_historical_data_test_profile_lr_sp_adult.csv')
df3 = pd.read_csv('historical_data/historical_data_train_profile_lr_sp_adult.csv')
df4 = pd.read_csv('historical_data/historical_data_test_profile_lr_sp_adult.csv')

col1 = df1.columns[-1]
col2 = df2.columns[-1]
col3 = df3.columns[-1]
col4 = df4.columns[-1]

plt.figure(figsize=(10, 6), dpi=300)
plt.scatter(df1[col1].values, df2[col2], label='Test data utility (F2)')
plt.scatter(df1[col1].values, df3[col3], color='r', label='Training data utility (P1)')
plt.scatter(df1[col1].values, df4[col4], color='g', label='Original Test data utility (F6)')

plt.xlabel("Similarity (P1_F2)")
plt.ylabel("Utility (fairness)")
plt.legend()
plt.tight_layout()
plt.show()
```
