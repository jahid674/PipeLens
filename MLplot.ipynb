#### Relaxed fgoal

```python
Grid_r=[2,3,4,13]
Grid_ES_r=[1,2,3,8]
l2c_r=[1,6,11,23]
BugDoc_r=[9,17,17,17]
PipeLens_r=[1,1,1,1]
```

```python

fsize = 20
params = {'legend.fontsize': fsize, 'legend.handlelength': 2}
plot.rcParams.update(params)

font = {'family': "sans serif", 'size': fsize}
matplotlib.rc('font', **font)

# --- Save path ---
folder_path = '/Users/hasan89/Documents/PipeLens_Plot/Fairness'
file_name = f"{task}_{case}.pdf"

os.makedirs(folder_path, exist_ok=True)
file_path = os.path.join(folder_path, file_name)

import os
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

# ... your existing code above ...

plt.figure(figsize=(6, 5))
plt.xticks(fontsize=fsize/1.2)
plt.yticks(fontsize=fsize/1.2)

#x_en = [0.84, 0.87, 0.94, 0.95]
x_en = [25, 50, 75, 100]

# Map to evenly spaced positions
labels = [f"{v:.0f}" for v in x_en]
xpos = np.arange(len(x_en))  # 0,1,2,3 (equally spaced)

# Plot against xpos (not x_en)
plt.plot(xpos, Grid_r,     marker='o', linestyle='-', color='navy',       markersize=13, label='Systematic')
plt.plot(xpos, Grid_ES_r,     marker='v', linestyle='-', color='m',       markersize=13, label='GridSearch-ES')
plt.plot(xpos, l2c_r,  marker='*', linestyle='-', color='red',        markersize=10, label='Learn2Clean')
plt.plot(xpos, BugDoc_r,   marker='d', linestyle='-', color='blueviolet', markersize=10, label='BugDoc')
plt.plot(xpos, PipeLens_r, marker='s', linestyle='-', color='black',      markersize=10, label='PipeLens')

#plt.xlabel('Goal Utility', labelpad=0, fontsize=fsize/1.2)
plt.xlabel('Percentile', labelpad=0, fontsize=fsize/1.2)
plt.ylabel('#iterations', labelpad=0, fontsize=fsize/1.2)
plt.legend(fontsize=fsize/1.4, loc='best', ncol=1)

# Equally spaced ticks with labels from x_en
ax = plt.gca()
ax.set_xticks(xpos)
ax.set_xticklabels(labels)
import numpy as np
from matplotlib.ticker import LogLocator, LogFormatterMathtext, NullFormatter
ys = [np.asarray(a) for a in [Grid_w, Grid_ES_w, l2c_w, BugDoc_w, PipeLens_w] if a is not None]
max_y = np.nanmax(np.concatenate(ys)) if ys else 0

ax = plt.gca()
ax.set_xticks(xpos)
ax.set_xticklabels(labels)

if max_y > 100:
    ax.set_yscale('log')
    ax.yaxis.set_major_locator(LogLocator(base=10))
    ax.yaxis.set_minor_locator(LogLocator(base=10, subs=np.arange(2,10)*0.1, numticks=10))
    ax.yaxis.set_major_formatter(LogFormatterMathtext(base=10))  # shows 10^n
    ax.yaxis.set_minor_formatter(NullFormatter())

plt.tight_layout()  
plt.savefig(file_path, format='pdf', dpi=1600, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
